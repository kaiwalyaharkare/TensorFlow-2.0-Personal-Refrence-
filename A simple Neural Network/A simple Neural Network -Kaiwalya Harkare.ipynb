Data preprocessing and common step wise procedure

```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

Read data from csv

```python
df = pd.read_csv('../DATA/fake_reg.csv')
```

```python
df.head() 
```

Convert data to NP array for tf

```python
X = df[['feature1','feature2']].values
y = df[['price']].values
```

Split data in to testing data and training data using test train spilt

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

Now we have to scale our data from values of zero to one using the MinMaxScaler this will turn the values of our data in values bettween one and two 

```python
from sklearn.preprocessing import MinMaxScaler
```

We will have to assign a object to the minmax scaler 

```python
scaler =  MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

            Now we start making the model i.e a simple Neural Network With a dense layer a layer in which all the neuron are conneted together

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
Model = Sequential()
```

```python
type(Model)
```

relu

 Choosing an optimizer and loss
Keep in mind what kind of problem you are trying to solve:

 For a multi-class classification problem
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


 For a binary classification problem
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
 For a mean squared error regression problem
model.compile(optimizer='rmsprop',
              loss='mse')

```python
model = Sequential()
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
# for final output we will not choose a activation fuction and choose only one nuron as we are doind time series data 
Model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

To train the model we use model.fit()

```python
model.fit(x=X_train,y=y_train,epochs=500)
```

```python
df_Loss = pd.DataFrame( model.history.history)
```
